```python
#from rdflib import Graph, Namespace, URIRef
#from rdflib.namespace import RDF, RDFS, XSD
#from SPARQLWrapper import SPARQLWrapper, JSON
import sqlite3
from src.helpers import *


# check
ttl_file = "machinery_kpi/ontologies/crawler_crane_ontology.ttl"  
rdf_graph = load_constraints(ttl_file)

# Define namespaces
EX = Namespace("urn:crawlercrane-ontology#")

# 2. Connect to the SQLite database and fetch telemetry data
conn = sqlite3.connect("logs_simulation/machine_data.db")
cursor = conn.cursor()

# Simulate an SQL query based on user input (adjust as needed)
query = """
SELECT id, timestamp, machine_id, component, sensor_name, sensor_value, sensor_unit 
FROM telemetry
where component = "urn:crawlercrane-ontology#Carbody" and machine_id = 1
limit 100
"""
rows = cursor.execute(query).fetchall()

# 3. Validate telemetry rows
valid_rows = []
invalid_rows = []

for row in rows:
    component_uri = component = row[3]
    #component_uri = EX[component] # uncomment when there are components names in the db wothout the namespace
    
    #print(f"Processing component: {component}")
    
    # Retrieve constraints
    constraints = get_constraints(rdf_graph, f"{component_uri}")
    #print(f"Constraints for {component}: {constraints}")
    
    # Validate row
    if check_constraints_for_row(row, constraints):
        valid_rows.append(row)
    else:
        invalid_rows.append(row)

print("Valid rows:")
for r in valid_rows:
    print(r)

print("Invalid rows:")
for r in invalid_rows:
    print(r)


conn.close()

```
